# 2. Axis

```python
import matplotlib.pyplot as plt
```

```python
x  = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y) 
plt.xlabel('x axis', color ='red', loc='right') # loc(location - left, center, right
plt.ylabel('y axis', color = 'blue', loc = 'top') # top, center, bottom
```

```python
plt.plot(x, y) 
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
