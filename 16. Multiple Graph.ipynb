# 16. Multiple Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('../Pandas//score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) # create 2x2 plots
fig.suptitle('putting multiple graphs')

# first graph
axs[0, 0].bar(df['name'], df['English'], label='English  score')
axs[0, 0].set_title('First Graph')
axs[0, 0].legend()
axs[0, 0].set(xlabel='Name', ylabel='score') # DIFFERENT 
axs[0, 0].set_facecolor('lightyellow') # background color

# second graph
axs[0, 1].plot(df['name'], df['English'], label='English  score')
axs[0, 1].plot(df['name'], df['Math'], label='Math  score')
axs[0, 1].set_title('Second Graph')
axs[0, 1].legend()
axs[0, 1].set(xlabel='Name', ylabel='score') # DIFFERENT 
axs[0, 1].set_facecolor('lightyellow') # background color

# third graph
axs[1, 0].barh(df['name'], df['height'], label='height')
axs[1, 0].set_title('Third Graph')
axs[1, 0].legend()
axs[1, 0].set(xlabel='Name', ylabel='height') # DIFFERENT 
axs[1, 0].set_facecolor('lightyellow') # background color

# first graph
axs[1, 1].plot(df['name'], df['Science'], label='score', color='g', alpha=0.5)
axs[1, 1].set_title('Forth Graph')
axs[1, 1].legend()
axs[1, 1].set(xlabel='Name', ylabel='score') # DIFFERENT 
axs[1, 1].set_facecolor('lightyellow') # background color
```
