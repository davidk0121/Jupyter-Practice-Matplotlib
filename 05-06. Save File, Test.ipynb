# 05. Save File

```python
import matplotlib.pyplot as plt
```

```python
x  = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.savefig('graph.png', dpi = 100)
```

```python
plt.figure(dpi = 200)
plt.plot(x, y)
plt.savefig('graph_200.png')
```

# 06. Text

```python
plt.plot(x, y, marker='o')

for idx,  txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt, ha = 'center')
```
