# 15. Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('../Pandas//score.xlsx')
df
```

```python
df['grade'] = [2, 3, 1, 4]
df
```

```python
plt.scatter(df['English'], df['Math'])
plt.xlabel('Eng score')
plt.ylabel('Math score')
```

```python
sizes = df['grade'] * 500
plt.scatter(df['English'], df['Math'], s=sizes)
plt.xlabel('Eng score')
plt.ylabel('Math score')
```

```python
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis', alpha=0.3)
plt.xlabel('Eng score')
plt.ylabel('Math score')
```

```python
plt.figure(figsize=(10,5)) # graph box size
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis', alpha=0.3)
plt.xlabel('Eng score')
plt.ylabel('Math score')
plt.colorbar(ticks=[1,2,3,4], label='grade') # bar on the right
```
