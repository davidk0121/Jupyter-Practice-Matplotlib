# 03. Legend 

```python
import matplotlib.pyplot as plt
```

```python
x  = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label = 'some data')
plt.legend() # use loc = ()
```

```python
plt.legend(ncol=3) # extend it to right
```

# 04. Style

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

## line color

```python
plt.plot(x, y, color='red')
```

## Graph size

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi = 50) # set dpi (size of the graph)
plt.plot(x, y)
```
