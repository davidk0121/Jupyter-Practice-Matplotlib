# 08. Bar Graph

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Sally', 'David', 'Daniel']
values = [158, 173, 175]

plt.bar(labels, values)
```

```python
labels = ['Sally', 'David', 'Daniel']
values = [158, 173, 175]

plt.bar(labels, values, color='g')
```

```python
labels = ['Sally', 'David', 'Daniel']
values = [158, 173, 175]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color = colors)
```

```python
labels = ['Sally', 'David', 'Daniel']
values = [158, 173, 175]

plt.bar(labels, values)
plt.ylim(150, 180) # set the y axis length
```

```python
plt.bar(labels, values, width = 0.5)
```

```python
plt.bar(labels, values, width = 0.5)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
```

# 09. Bar Graph - 2

```python
labels = ['Sally', 'David', 'Daniel']
values = [158, 173, 175]

plt.barh(labels, values)
plt.xlim(150, 180) 
```

```python
bar = plt.bar(labels, values)
plt.ylim(150, 180) 

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='blue')
```
